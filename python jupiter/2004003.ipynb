```python
import numpy as np
import matplotlib.pyplot as plt

Samping_rate = 5000  
Time = 0.025
t_continuous = np.arange(0, Time, 1/Samping_rate)
x_discrete = 3 * np.cos(2000 * np.pi * t_continuous) + 5 * np.sin(6000 * np.pi* (t_continuous - 2)) + 10 * np.cos(1200 * np.pi * (t_continuous + 2))

plt.plot(t_continuous, x_discrete, 'bo-', label='Discrete Signal')
plt.title('Discrete Time Signal After Sampling Rate Samping_rate = 5000')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```
