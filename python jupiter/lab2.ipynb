```python
import numpy as np
import matplotlib.pyplot as plt

x = [-1, 2, 3,-5, 7, 8]
l = len(x)

origin = 2
n = np.arange(-origin, l-origin)
print(n)
plt.figure(figsize = (20,20))
plt.subplot(2,3,1)
plt.axhline(y=0, c="black")
plt.plot(n, x, 'r*')
plt.subplot(2,3,2)
plt.axhline(y=0, c="black")
n = np.arange(-origin+2, l-origin+2)
plt.plot(n, x, 'b>')
# plt.plot(n+2, x, 'b>')
plt.subplot(2,3,3)
plt.axhline(y=0, c="black")
n = np.arange(-origin-3, l-origin-3)
plt.plot(n, x, 'y<')
plt.subplot(2,3,4)
xx = x[::-1]
n = np.arange(-(l-origin-1), origin+1)
plt.axhline(y=0, c="black")
plt.plot(n, xx, 'g^')
y = x[::2]
ny = np.arange(-origin, l-origin, 2)
plt.subplot(2, 3, 5)
plt.plot(ny, y, 'r*')
n = np.arange(-origin, l-origin)
plt.subplot(2,3,6)
plt.axhline(y=0, c="black")
plt.plot(n, x, 'y<')
```
