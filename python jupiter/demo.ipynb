```python
import numpy as np
import matplotlib.pyplot as plt
draw=8000
sampling_rate1=8000

frequency_1=12
frequency_2=5

sampling_rate2=7

x1=np.arange(0, draw, draw/sampling_rate1)
x2=np.arange(0, draw, draw/sampling_rate1)
x3=np.arange(0, draw, draw/sampling_rate2)

y1=np.cos(2*np.pi*frequency_1*x1/sampling_rate1)
y2=np.cos(2*np.pi*frequency_2*x2/sampling_rate1)
y3=np.cos(2*np.pi*frequency_2*x3/sampling_rate1)

plt.plot(x1,y1, color='red')
plt.plot(x2,y2, color='blue')
plt.scatter(x3,y3, color='black')

plt.title("Sin Graph")
plt.xlabel('sampling(10000)')
plt.ylabel('voltage(V)')
plt.show()
```
