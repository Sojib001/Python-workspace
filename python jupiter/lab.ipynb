```python
import numpy as np
import matplotlib.pyplot as plt
x = [-1, 2, 3, -5, 7, 8]
origin = 2
n = np.arange(-origin, len(x) - origin)
print(x)
plt.figure(figsize=(18, 8))
plt.subplot(2, 3, 1)
plt.axhline(y = 0, c = 'green')
plt.plot(n , x, 'r')
plt.title("Original")
plt.subplot(2, 3, 2)
plt.axhline(y = 0, c = 'black')
plt.plot(n - 2, x, 'b')
plt.title("System shifted")
plt.subplot(2, 3, 3)
plt.axhline(y = 0, c = 'black')
plt.plot(n - 2, x, 'b')
plt.title("Input shifted")

plt.subplot(2, 3, 4)
plt.axhline(y = 0, c = 'black')
plt.plot(n , n*x, 'b')
plt.title("Original")

plt.subplot(2, 3, 5)
plt.axhline(y = 0, c = 'black')
plt.plot(n - 2, (n-2)*x, 'b')
plt.title("System shifted")

plt.subplot(2, 3, 6)
plt.axhline(y = 0, c = 'black')
plt.plot(n - 2, n*x, 'b')
plt.title("Input shifted")


```
